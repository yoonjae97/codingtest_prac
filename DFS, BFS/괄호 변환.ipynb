```python
def balanced_str(p):
    count = 0
    for i in range(len(p)):
        if p[i] == '(':
            count += 1
        else:
            count -= 1
    
        if count == 0:
            return i

def right_str(p):
    count = 0
    for i in p:
        if i == '(':
            count += 1
        else:
            if count == 0 :
                return False
            count -= 1
    return True

def reverse_str(p):
    for i in range(len(p)):
        if p[i] == '(':
            p[i] = ')'
        else:
            p[i] = '('
    return p

def solution(p):
    answer = ''
    if p == '':
        return answer
    
    index = balanced_str(p)
    
    u = p[:index + 1]
    v = p[index + 1:]

    if right_str(u):
        answer = u + solution(v)
    
    else:
        answer = '('
        answer += solution(v)
        answer += ')'
        u = list(u[1:-1])
        u = reverse_str(u)
        answer += "".join(u)
    return answer
```
