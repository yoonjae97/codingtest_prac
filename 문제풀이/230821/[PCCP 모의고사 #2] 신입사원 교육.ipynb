```python
# 수정된 코드
import heapq
def solution(ability, number):
    heapq.heapify(ability)
    for i in range(number):
        total = 0

        total += heapq.heappop(ability)
        total += heapq.heappop(ability)
       
        heapq.heappush(ability, total)
        heapq.heappush(ability, total)
        
    return sum(ability)

# 참고자료 https://velog.io/@soorim_yoon/PCCP-%EC%8B%A4%EC%A0%84%EB%AA%A8%EC%9D%98%EA%B3%A0%EC%82%AC-202-%EC%8B%A0%EC%9E%85%EC%82%AC%EC%9B%90-%EA%B5%90%EC%9C%A1
# 참고 자료2 https://kjhoon0330.tistory.com/entry/Python-heapq-%EB%AA%A8%EB%93%88
# 참고 자료3 및 그림 출처 https://brownbears.tistory.com/550

# heap 자료구조는 이진트리의 형태로 min heap 구조일 경우 제일 작은 값을 트리의 최상단에 위치시킨다.
# 따라서 heappop함수를 통해 제일 작은 값을 출력하게 되는 것이다.
# 하지만 sorted 함수의 정렬과는 구조가 다르기 때문에 heappop이 아닌 인덱싱을 통해 값을 불러올 경우 2번째 인덱스의 값이
# 반드시 2번째로 작은 값일 것이라는 보장은 없으므로 주의가 필요하다.
```
