```python
from collections import deque

def solution(places):
    result_lst = []
    cnt = 0
    for place in places:
        p_lst = []
        val = False
        cnt += 1
        for i in range(5):
            for j in range(5):
                if place[i][j] == 'P':
                    p_lst.append([i, j])
                 
        for x, y in p_lst:
            visit = [[False] * 5 for i in range(5)]  
            distance = [[0] * 5 for i in range(5)]  
            
                        
            q = deque()            
            q.append([x, y])
            visit[x][y] = True            

            while q:
                ax, ay = q.popleft()
                # 상 하 좌 우
                dx = [-1, 1, 0, 0]
                dy = [0, 0, -1, 1]

                for i in range(4):
                    nx = ax + dx[i]
                    ny = ay + dy[i]

                    if nx > -1 and nx < 5 and ny > -1 and ny < 5 and visit[nx][ny] == False:
                        
                        if place[nx][ny] != 'O':
                            q.append([nx, ny])
                            visit[nx][ny] = True
                            distance[nx][ny] = distance[ax][ay] + 1
                        
                        if place[nx][ny] == 'P' and abs(x - nx) + abs(y - ny) <= 1:
                            val = True
                            break
            #print(cnt, '/', x, y, distance)          
        if val == True:
            result_lst.append(0)
        else:
            result_lst.append(1)
                    
    return (result_lst)
```

```python
from collections import deque

def bfs(p):
    start = []
    
    for i in range(5): # 시작점이 되는 P 좌표 구하기
        for j in range(5):
            if p[i][j] == 'P':
                start.append([i, j])
    
    for s in start:
        queue = deque([s])  # 큐에 초기값
        visited = [[0]*5 for i in range(5)]   # 방문 처리 리스트
        distance = [[0]*5 for i in range(5)]  # 경로 길이 리스트
        visited[s[0]][s[1]] = 1
        
        while queue:
            y, x = queue.popleft()
        
            dx = [-1, 1, 0, 0]  # 좌우
            dy = [0, 0, -1, 1]  # 상하

            for i in range(4):
                nx = x + dx[i]
                ny = y + dy[i]

                if 0<=nx<5 and 0<=ny<5 and visited[ny][nx] == 0:
                    
                    if p[ny][nx] == 'O':
                        queue.append([ny, nx])
                        visited[ny][nx] = 1
                        distance[ny][nx] = distance[y][x] + 1
                    
                    if p[ny][nx] == 'P' and distance[y][x] <= 1:
                        return 0
    return 1


def solution(places):
    answer = []
    
    for i in places:
        answer.append(bfs(i))
    
    return answer
```
